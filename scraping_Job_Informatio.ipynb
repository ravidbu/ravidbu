# Monster

```python
import requests
from bs4 import BeautifulSoup
import xlwt
```

```python
file = xlwt.Workbook() 
table = file.add_sheet('Monster') 
table_row = 0 
```

```python
URL = 'https://www.monster.com/jobs/search/?q=Business-Analyst&where=Boston__2C-MA'
page = requests.get(URL)
soup = BeautifulSoup(page.content, 'html.parser')
results = soup.find(id='ResultsContainer')
#print(results.prettify())
job_elems = results.find_all('section', class_='card-content')
```

```python
for job_elem in job_elems:
    
    title_elem = job_elem.find('h2', class_='title')
    company_elem = job_elem.find('div', class_='company')
    location_elem = job_elem.find('div', class_='location')

    if None in (title_elem, company_elem, location_elem):
        continue
        
    print(title_elem.text.strip())
    print(company_elem.text.strip())
    print(location_elem.text.strip())
    print()
    
    table.write(table_row, 0, title_elem.text.strip()) 
    table.write(table_row, 1, company_elem.text.strip())
    table.write(table_row, 2, location_elem.text.strip()) 
    table_row = table_row + 1 

#file.save('Position_Information.xlsx')
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
#import selenium
#from selenium import webdriver
#import time
```

```python
base_url = 'http://www.indeed.com/jobs?q=business+analyst&jt=fulltime&sort='
sort_by = 'date'          # sort by date
start_from = '&start='    # start page number

pd.set_option('max_colwidth',500)    # to remove column limit (Otherwise, we'll lose some info)
df = pd.DataFrame()   # create a new data frame

for page in range(1,11): # page from 1 to 10
    page = (page-1) * 10  
    url = "%s%s%s%d" % (base_url, sort_by, start_from, page) # get full url 
    page = requests.get(url)
    target = BeautifulSoup(page.content, 'html.parser')
    results = target.find(id='resultsCol')
    BA_jobs = results.find_all('div', class_='title')  
    for BA_jobs_link in BA_jobs:
        link = BA_jobs_link.find('a')['href']
        links = 'http://www.indeed.com/' + link
        website = requests.get(links)
        SOUP = BeautifulSoup(website.content, 'html.parser')
        job_title = SOUP.find('h3',{'class':'icl-u-xs-mb--xs icl-u-xs-mt--none jobsearch-JobInfoHeader-title'})
        job_link = links
        jd = SOUP.find('div',{'class':'jobsearch-jobDescriptionText'})

        df = df.append({'job_title': job_title.text.strip(),'job_link': job_link,'job_information': jd.text.strip()}, ignore_index=True)

        
job_data = df.to_csv("job_data.csv")

        #the way of using selenium to get website
        #chr = webdriver.Chrome('/Users/guorong/Downloads/chromedriver')
        #text = chr.get(links)
        #time.sleep(5)
        #jd = chr.find_element_by_class_name("jobsearch-jobDescriptionText")
        #jd.text.strip()
        
  
```
